```python
import numpy as np

print('NumPy version:', np.__version__)
```

```python
# From a Python list

a = np.array([1, 2, 3, 4, 5])

print('Array:', a)
print('Shape:', a.shape)    # (5,) — 1-D, 5 elements
print('Dtype:', a.dtype)    # int64 by default
print('Ndim:', a.ndim)      # 1
```

```python
# 2-D array (matrix)

mat = np.array([[1, 2, 3],
                [4, 5, 6]])

print('Shape:', mat.shape)   # (2, 3) — 2 rows, 3 cols
print(mat)
```
